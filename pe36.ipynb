```python
list=[]
for i in range(1, 1000000):
    if(str(i)==str(i)[::-1]):
#         print("The string is a palindrome")
        list.append(i)
print(list)
```

```python
list2=[]
for x in list:
    list2.append(bin(x).replace("0b", ""))
```

```python
print(list2)
```

```python
list3=[]
for i in list2:
    if(str(i)==str(i)[::-1]):
#         print("The string is a palindrome")
        list3.append(i)
print(list3)
```

```python
def binaryToDecimal(n):
    return int(n,2)
```

```python
list4=[]
for y in list3:
    list4.append(binaryToDecimal(y))
print(list4)
```
