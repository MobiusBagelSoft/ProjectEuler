```python
import math
from tqdm import tqdm
```

```python
def d(k):
    res = [(int(x)) for x in str(k)]
    return sum(list(res))
```

```python
d(42)
```

```python
dict={}
def d2(k):
    if k in dict.keys():
        return dict.get(k)
    else:
        output=d(k)
        dict.update({k: output})
        return output
```

```python
d2(42)
```

```python
def check(k):
    res = [(int(x)) for x in str(k)]
    res[-1]*7
    return sum(list(res))
```

```python
def s(n):
    total=0
    bound=10**n
    for x in tqdm(range(bound)):
        if x % 23 == 0 and d2(x)==23:
            total=(total+1) % 10**9
    return total
```

```python
s(9)
```
