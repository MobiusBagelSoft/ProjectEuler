```python
from sympy.ntheory.factor_ import totient
from tqdm import tqdm
```

```python
sorted(str(123))==sorted(str(321))
```

```python
s=set()
for n in tqdm(range(2, 10**7)):
#     a=totient(n)
    if sorted(str(n))==sorted(str(totient(n))):
        s.add(n)
# print(min(s))
```
