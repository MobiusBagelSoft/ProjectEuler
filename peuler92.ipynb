```python
from tqdm import tqdm
import math
```

```python
def square(y):
    res = [int(x)**2 for x in str(y)]
    return sum(list(res))
```

```python
dict={}
def square2(n):
    if n in dict.keys():
        return dict.get(n)
    else:
        output=square(n)
        dict.update({n: output}) 
        return output
    
```

```python
s=set()
total=0
for n in tqdm(range(1,10**7)):
    result=n
    s.add(result)
    while 89 not in s:
        result=square2(result)
        s.add(result)
        if 1 in s:
            s=set()
            break
    if 89 in s:
        total=total+1
        s=set()
    

print(total)
```
